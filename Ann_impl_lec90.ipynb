Basically one-hot-encoding transform string data into integer data form.

Imagine a bank has customers from different places (Geography) and genders (Male/Female). One-hot encoding helps turn these categories into a format a computer can understand for building a model to predict churn (leaving the bank).

Before One-Hot Encoding:

Geography: The data might have text entries like "France", "Germany", "Spain".
Gender: It could be "Male" or "Female".
The computer struggles with these because it prefers numbers. One-hot encoding creates a new set of columns for each category.

After One-Hot Encoding:

New Columns: There will be separate columns for each geography (e.g., France, Germany, Spain) and gender (Male, Female).
Values: Each new column will have values of 0 or 1.
Interpretation:
A customer from France will have 1 in the "France" column and 0 in all other geography columns.
A male customer will have 1 in the "Male" column and 0 in the "Female" column.
This way, the computer can understand the customer's location and gender using these new columns with 0s and 1s, making it easier to analyze if these factors influence customer churn.

Benefit: One-hot encoding helps machine learning models treat categories like gender and geography effectively for tasks like churn prediction.

````python
#the above all are categorical features. Therefore now we use ine-hot encoding.

#one-hot encoding
df=pd.get_dummies(df, columns=["Geography", "Gender"], drop_first=True)


#female=1; male=0
#france=(1,0,0); spain=(0,1,0); germany=(0,0,1)
````

Sequential groups a linear stack of layers into a tf.keras.Model.

Sequential provides training and inference features on this model.

Maintain layers working in a sequential manner and also maintain the forward and backward propogation.

````python
model=Sequential()

#dence is responsible for adding layers.

model.add(Dense(11,activation="sigmoid",input_dim=11))#input layer
model.add(Dense(11,activation="sigmoid"))#hidden layer:- inside this hidden layer we can take any number of dence but with less than 11.
model.add(Dense(1,activation="sigmoid"))#output layer
#there are tota 11 columns in the dataset and hence 11 input therefore 11 numbr of neurons and we need a single output therefore 1 neuron for the output.
````

````python
#splliting the data into train and test

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
````

Standard scaling is a technique used to preprocess data before feeding it into a machine learning model. Imagine you have a bunch of ingredients for a recipe, but some are measured in cups and others in grams. It's hard to follow the recipe if everything isn't using the same measurement system. Standard scaling does something similar for machine learning models.

````python
#in Neural network we use gradient decent optimizer and GD uses standard scaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
````

````python
# now we have to compile the model with the optimizer
#more follow tenserflow documnetation

''''
model.compile(optimizer=, loss=, metrics=)
'''
model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])       #since we have a classification problem and hence we use loss=binary_crossentropy
                                                                        #here accuracy is training accuracy not the predicted or final accuracy
````

````python
model.fit(X_train,y_train,batch_size=50,epochs=10,verbose=1)
````

````python
history=model.fit(X_train_scale,y_train,batch_size=50,epochs=10,verbose=1)

'''
In machine learning, an epoch refers to one complete pass through the entire training dataset.
 It's like giving your machine learning model one big run-through of all the training data it has available.

'''


'''
10 Epochs = 10 Training Runs: With epochs=10, the model will go through the training data set a total of 10 times,
 updating its internal parameters (like weights and biases) each time to improve its performance.

'''


'''
verbose=1 controls how much information you see during the training process of your model.

'''

'''

when training models with TensorFlow or similar libraries,
 batch_size=50 refers to a hyperparameter that determines the number of data points used in one update of the model's internal parameters during training.
 batch_size=50 --> 50dataset for each epoch.

'''
````
